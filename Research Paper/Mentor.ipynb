```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "college.csv"  # Ensure the file is in the working directory
df = pd.read_csv(file_path)

# List of theory and practical subjects for Semester 1
sem1_theory = ["Math-1 Theory", "Physics Theory", "Java-1 Theory", "Software Engineering Theory", "Environmental Science Theory"]
sem1_practical = ["Physics Practical", "Java-1 Practical", "Software Engineering Practical", "IOT Workshop Practical", "Computer Workshop Practical"]
sem1_attendance = ["Math-1 Attendance", "Physics Attendance", "Java-1 Attendance", "Software Engineering Attendance", "Environmental Science Attendance", "IOT Workshop Attendance"]

# Compute average attendance, theory marks, and practical marks per student for Sem 1
df['Avg_Attendance_Sem1'] = df[sem1_attendance].mean(axis=1)
df['Avg_Theory_Sem1'] = df[sem1_theory].mean(axis=1)
df['Avg_Practical_Sem1'] = df[sem1_practical].mean(axis=1)

# Perform ANOVA to check if mentor assignment affects performance
anova_attendance = stats.f_oneway(*[df[df['Mentor-1'] == mentor]['Avg_Attendance_Sem1'].dropna() for mentor in df['Mentor-1'].unique()])
anova_theory = stats.f_oneway(*[df[df['Mentor-1'] == mentor]['Avg_Theory_Sem1'].dropna() for mentor in df['Mentor-1'].unique()])
anova_practical = stats.f_oneway(*[df[df['Mentor-1'] == mentor]['Avg_Practical_Sem1'].dropna() for mentor in df['Mentor-1'].unique()])

print("ANOVA Results Sem-1:")
print("Attendance: p-value =", anova_attendance.pvalue)
print("Theory: p-value =", anova_theory.pvalue)
print("Practical: p-value =", anova_practical.pvalue)

# Visualization
plt.figure(figsize=(15, 5))
sns.boxplot(x='Mentor-1', y='Avg_Attendance_Sem1', data=df)
plt.xticks(rotation=45)
plt.title("Attendance Distribution Across Mentors")
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(x='Mentor-1', y='Avg_Theory_Sem1', data=df)
plt.xticks(rotation=45)
plt.title("Theory Marks Distribution Across Mentors")
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(x='Mentor-1', y='Avg_Practical_Sem1', data=df)
plt.xticks(rotation=45)
plt.title("Practical Marks Distribution Across Mentors")
plt.show()




# List of theory and practical subjects for Semester 2
sem2_theory = ["Math-2 Theory", "Fundamental of Electronics and Electrical Theory", 
               "Java-2 Theory", "Data Structures using Java Theory", "DBMS Theory"]

sem2_practical = ["Fundamental of Electronics and Electrical Practical", "Java-2 Practical", 
                  "Data Structures using Java Practical", "DBMS Practical"]
sem2_attendance = ["Math-2 Attendance", "Fundamental of Electronics and Electrical Attendance", "Java-2 Attendance", 
                   "Data Structures using Java Attendance", "DBMS Attendance"]

# Compute average attendance, theory marks, and practical marks per student for Sem 2
df['Avg_Attendance_Sem2'] = df[sem2_attendance].mean(axis=1)
df['Avg_Theory_Sem2'] = df[sem2_theory].mean(axis=1)
df['Avg_Practical_Sem2'] = df[sem2_practical].mean(axis=1)

# Perform ANOVA to check if mentor assignment affects performance
anova_attendance2 = stats.f_oneway(*[df[df['Mentor-2'] == mentor]['Avg_Attendance_Sem2'].dropna() for mentor in df['Mentor-2'].unique()])
anova_theory2 = stats.f_oneway(*[df[df['Mentor-2'] == mentor]['Avg_Theory_Sem2'].dropna() for mentor in df['Mentor-2'].unique()])
anova_practical2 = stats.f_oneway(*[df[df['Mentor-2'] == mentor]['Avg_Practical_Sem2'].dropna() for mentor in df['Mentor-2'].unique()])

print("ANOVA Results Sem-2:")
print("Attendance: p-value =", anova_attendance2.pvalue)
print("Theory: p-value =", anova_theory2.pvalue)
print("Practical: p-value =", anova_practical2.pvalue)

# Visualization
plt.figure(figsize=(15, 5))
sns.boxplot(x='Mentor-2', y='Avg_Attendance_Sem2', data=df)
plt.xticks(rotation=45)
plt.title("Attendance Distribution Across Mentors")
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(x='Mentor-2', y='Avg_Theory_Sem2', data=df)
plt.xticks(rotation=45)
plt.title("Theory Marks Distribution Across Mentors")
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(x='Mentor-2', y='Avg_Practical_Sem2', data=df)
plt.xticks(rotation=45)
plt.title("Practical Marks Distribution Across Mentors")
plt.show()
```
